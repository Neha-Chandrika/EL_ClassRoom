```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
