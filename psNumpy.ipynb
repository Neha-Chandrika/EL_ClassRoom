```python
#Elementary-wise operations with two arrays

x = np.array([1,2,3])
y = np.array([10,20,30])

print("the x array " , x)
print("the y array " , y)
print(x + y)
print(x - y)
print(x / y)
print(x % y)
print(x ** y)
print(x * y)
```

```python
#very imp 2 questions

data = np.array([10, 20, 30, 40, 50])

print(data.mean()) #average
print(data.sum()) #total sum
print(data.min()) #minimum
print(data.max()) #maximum
print(data.std()) #standard deviation
#till here mam
print(data.var())
print(data.cumsum())
print(data.cumprod())
```

```python
#2D array (matrices)  and indexing, Creating, accessing elements
m = np.array([[1,2,3],
             [4,5,6]])

print(m[0,0]) #fisrt row, first column -> 1
print(m[1,2]) #second row, third column


#first row
print(m[0,:])

#first column
print(m[:,0])
```

```python
#mathematical functions

a = np.array([[1,2],
             [3,4]])
b = np.array([[5,6],
              [7,8]])

print("Add: ", np.add(a,b))
print("Subtract:", np.subtract(a,b))
print("Multiply:", np.multiply(a,b))
print("Divide: ",np.divide(a,b))
print("Square Root:", np.sqrt(a))
print("Power:", np.power(a,3))
```

```python
#linear
x = np.array([1,2,3])
y = np.array([10,20,30])

print(np.dot(x,y)) #dot product between two vector
```

```python
m = np.array([[1,2],
             [4,5]])

print(np.linalg.inv(m)) #inverse of a matrix
#NOTE : Last 2 dimensions of the array must be square
```

```python
#creating special arrays , NumPy can create without starting from lists

zeroes = np.zeros((2,3)) # 2x3 array of 0s
print(zeroes)
ones = np.ones((3,3))
print(ones)
range_arr = np.arange(0,10,2)
print(range_arr)
linspace_arr = np.linspace(0,1,5)
print(linspace_arr)
```
